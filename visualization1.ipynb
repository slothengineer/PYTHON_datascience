```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("this is x vs y graph")
```

```python
plt.scatter(x,y,c="g")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("this is x vs y graph")
```

```python
plt.scatter(x,y,c="#FF8D33",alpha=.5, marker="x") #c=color #alpha=intensity of color
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("this is x vs y graph")
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x,y,c="#FF8D33",alpha=1, marker="x")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("this is x vs y graph")
plt.grid()
```
